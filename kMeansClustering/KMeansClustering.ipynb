### KMeansClustering - Grouping Data

```python
#Required Imports
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
#Creating Data
data_points, groups = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)
plt.scatter(data_points[:,0],data_points[:,1])
```

```python
# After Applying The Clustering Alg Our Data Should Look Something Like This.
colors = ["b", "c", "k", "m", "g"]
for i in range(len(data_points)):
    plt.scatter(data_points[i][0],data_points[i][1], color=colors[groups[i]])
```

```python
#Lets Import The KMeans Alg And Group The Data
from sklearn.cluster import KMeans
pred = KMeans(n_clusters=4, random_state=0).fit_predict(data_points)
print(pred)
```

```python
#Lets Plot The Same Graph Using The Predictions From KMeans
for i in range(len(data_points)):
    plt.scatter(data_points[i][0],data_points[i][1], color=colors[pred[i]])
```
